```python
def thisyear(time_obj):
    current_time = time_obj.strftime("%d-%b-%Y") #functions related to year, month and date for automating entering current date and month
    return(current_time[7:11]) 
    
```

```python
def beforeyear(time_obj):
    cur_month = time_obj.month
    sixmonth_before_urmonth = (cur_month- 6)%12
    if (cur_month <= 6):
        before_year = str(time_obj.year-1)
        return before_year
    else:
        before_year = str(time_obj.year)
        return before_year
```

```python
def beforemonth(time_obj):
    cur_month = time_obj.month
    sixmonth_before_curmonth = (cur_month- 6)%12
    
   
    if(sixmonth_before_curmonth==1):
        return 'Jan'
    elif(sixmonth_before_curmonth==2):
        return 'Feb'
    elif(sixmonth_before_curmonth==3):
        return 'Mar'
    elif(sixmonth_before_curmonth==4):
        return 'Apr'
    elif(sixmonth_before_curmonth==5):
        return 'May'
    elif(sixmonth_before_curmonth==6):
        return 'June'
    elif(sixmonth_before_curmonth==7):
        return 'July'
    elif(sixmonth_before_curmonth==8):
        return 'Aug'
    elif(sixmonth_before_curmonth==9):
        return 'Sep'
    elif(sixmonth_before_curmonth==10):
        return 'Oct'
    elif(sixmonth_before_curmonth==11):
        return 'Nov'
    else:
        return 'Dec'
```

```python
def thismonth(time_obj):
    current_time = time_obj.strftime("%d-%b-%Y")
    if(current_time[3:6]=='Jun'):
        return 'June'
    elif(current_time[3:6]=='Jul'):
        return 'July'
    else:
        return(current_time[3:6])
```

```python
def date(time_obj):
    current_time = time_obj.strftime("%d-%b-%Y")
    return(current_time[:2])
  
```

```python
def info(dataset):  #function to display different types of data visualization of the dataframe sent as argument
    line=dataset.iplot(x = 'Date', y= ['Prev Close'], xTitle = 'Time',yTitle = 'Price', title= 'LINE PLOT OF HISTORIC PREVIOUS CLOSE PRICE OF CHOSEN STOCK')
    display(line)
    #line plot
    
    dataset.iplot(kind = 'area', x='Date', y='Turnover',fill = True, title= 'AREA PLOT OF TURNOVER WRT YEAR', xTitle='Time',yTitle='Turnover')
    #area plot
    
    dataset[['Open','Close']].iplot(kind = 'hist', bins =25,barmode='group',bargap = 0.5, xTitle='Groups of values', yTitle='Number of values', title='GROUPING OF VALUES OF HISTORIC OPEN AND CLOSE PRICES OF CHOSEN STOCK USING HISTOGRAM')
    #histogram
    
    dataset[['Prev Close','Last','VWAP']].iplot(kind= 'box', xTitle='Columns',yTitle='Price',title='BOX PLOT OF PREV-CLOSE, LAST & VWAP VALUES OF CHOSEN STOCK HISTORICALLY') 
    #Box plot
    
    dataset.iplot(kind = 'bar',x='Date', y=['Open','High','Low','Close'],barmode='stack',bargap = 0.4,xTitle = 'Time',yTitle = 'Price',title='STACKED BAR PLOT OF OPEN,HIGH,LOW,CLOSE (HISTORIC PRICES)')
    #Stacked bar plot
    
    fig = px.scatter_3d(dataset,x='Open',y='Close',z='Volume',color='Volume',opacity=0.7)
    fig.update_layout(title="3D SCATTER OF OPEN, CLOSE(HISTORIC) WRT VOLUME")
    fig.show()
    #3D scatter plot
    
```
